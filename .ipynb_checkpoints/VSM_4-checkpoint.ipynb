```python
from collections import Counter

from scipy.stats import chi2_contingency
import statistics as stats
import scipy.linalg as la

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import normalize

np.set_printoptions(linewidth=np.inf)
```

```python
# Function to read file, returns title and text
def readFile(name):
    with open(name) as file:
        name = file.readline()
        buff = file.readline()
        file.close()
        
    return name, buff
    
name_x, text_x = readFile('data/011.txt')
name_y, text_y = readFile('data/008.txt')
    
text_y[:50]
```

```python
text_x_len = len(text_x)

# Text x character counts
text_x_count = dict(sorted(Counter(text_x).items()))
# Text x character probabilities, counts normalized by text length
text_x_norm = {key: val / text_x_len for key, val in text_x_count.items()}
```

```python
text_y_len = len(text_y)

# Text y character counts
text_y_count = dict(sorted(Counter(text_y).items()))
# Text y character probabilities, counts normalized by text length
text_y_norm = {key: val / text_y_len for key, val in text_y_count.items()}
```

```python
def plotGraph(text_x_data, text_y_data, title):
    width = 0.40

    fig, ax = plt.subplots(figsize=(14, 6))
    ax.bar(np.arange(len(text_x_data)) - width/2, text_x_data.values(), width, label = name_x, color='lightskyblue')
    ax.bar(np.arange(len(text_y_data)) + width/2, text_y_data.values(), width, label = name_y)

    ax.set_title(title, fontsize=12)

    plt.xticks(np.arange(len(text_x_data)), text_x_data.keys(), fontsize=11)
    plt.yticks(fontsize=11)
    plt.legend()

    plt.show()
    
plotGraph(text_x_count, text_y_count, 'Character counts')
```

## Úloha 1
Za předpokladu výše odhadněte matici přechodu markovského řetězce pro první text. Pro odhad matice přechodu vizte přednášku 17. Odhadnuté pravděpodobnosti přechodu vhodně graficky znázorněte, např. použitím heatmapy.

```python
# Get index of character in transition matrix, 0 corresponds to space, then follow letters
def getIndex(letter):
    ASCII_BEG = 97
    index = ord(letter) - ASCII_BEG + 1
    
    if index < 0:
        return 0
    
    return index

# Build transition matrix from text
def transitionMatrix(text):
    P = np.matrix(np.zeros((len(text_x_count), len(text_x_count)), dtype = np.int))
    
    for index, letter in enumerate(text):
        index_from = getIndex(letter)
        
        if (index + 1 < len(text)) :
            index_to = getIndex(text[index + 1])
            P[index_from, index_to] += 1
    
    return P
```

```python
P_x = transitionMatrix(text_x)
print(P_x)
```

```python
P_x = np.asmatrix(normalize(np.asarray(P_x), axis=1, norm='l1'))
```

```python
# Function to plot a heatmap
def plotHeatmap(P):
    fig, ax = plt.subplots(figsize=(12,10))
    ax = sns.heatmap(P, cmap="mako", xticklabels=list(text_x_count.keys()), yticklabels=list(text_x_count.keys()))
    plt.show()
    
plotHeatmap(P_x)
```

## Úloha 2
Na základě matice z předchozího bodu najděte stacionární rozdělení 
π tohoto řetězce pro první text.

```python
# Find stationary distribution for given transition matrix
def stationaryDistribution(P):
    M = np.transpose(P - np.eye(len(text_x_count)))
    
    pi = np.transpose(la.null_space(M)) 
    pi = normalize(np.asarray(pi), axis=1, norm='l1')
    
    return pi
    
pi_x = stationaryDistribution(P_x)   
pi_x
```

```python
# Just to make sure the normalized probabilities sum up to one
np.sum(pi_x)
```

```python
# Just to make sure the vector is in fact stationary
pi_x * P_x
```

## Úloha 3
Porovnejte rozdělení znaků druhého textu se stacionárním rozdělením 
π, tj. na hladině významnosti 5 % otestujte hypotézu, že rozdělení znaků druhého textu se rovná rozdělení
π z předchozího bodu.

```python
# Function to create a new dictionary, keys to be merged specified in argument
def mergeColumns(dictionary, new_key, to_merge):
    dictionary_merged = dictionary.copy()
    
    merged = 0
    for key in to_merge:
        merged += dictionary_merged.get(key)
        dictionary_merged.pop(key)

    dictionary_merged.update({new_key:merged})

    return dictionary_merged
```

```python
# Find character counts from stationary distribution
pi_x_counts = []

for index in range(0, pi_x.shape[1]):
    pi_x_counts.append(round(pi_x[0][index] * len(text_x)))
    
print(pi_x_counts)
```

```python
# Create dict for easier merging
text_x_pi = dict()

for index, key in enumerate(text_y_count.keys()):
    text_x_pi[key] = pi_x_counts[index]
    
print(text_x_pi)
```

```python
text_x_merged = mergeColumns(text_x_pi, 'qxz', ['q', 'x', 'z'])
print(text_x_merged)
```

```python
chars_obs_x = list(text_x_merged.values())
print(chars_obs_x)
```

```python
text_y_merged = mergeColumns(text_y_count, 'qxz', ['q', 'x', 'z'])
print(text_y_merged)
```

```python
chars_obs_y = list(text_y_merged.values())
print(chars_obs_y)
```

```python
chars_obs = np.array([chars_obs_x, chars_obs_y])
chars_obs
```

```python
# Chi-square test of independence of variables in a contingency table
s, p, dof, expected = chi2_contingency(chars_obs)
s
```
