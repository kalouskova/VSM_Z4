```python
import scipy.stats
import numpy as np

from collections import Counter
import matplotlib.pyplot as plt
import operator

from dahuffman import HuffmanCodec
```

## Úloha 1
Z obou datových souborů načtěte texty k analýze. Pro každý text zvlášť odhadněte pravděpodobnosti znaků (symbolů včetně mezery), které se v textech vyskytují. Výsledné pravděpodobnosti graficky znázorněte.

```python
# Function to read file, returns title and text
def readFile(name):
    with open(name) as file:
        name = file.readline()
        buff = file.readline()
        file.close()
        
    return name, buff
    
name_x, text_x = readFile('data/011.txt')
name_y, text_y = readFile('data/008.txt')
    
text_y[:50]
```

```python
text_x_len = len(text_x)

# Text x character counts
text_x_count = dict(sorted(Counter(text_x).items()))
# Text x character probabilities
text_x_norm = {key: val / text_x_len for key, val in text_x_count.items()}
```

```python
text_y_len = len(text_y)

text_y_count = dict(sorted(Counter(text_y).items()))
text_y_norm = {key: val / text_y_len for key, val in text_y_count.items()}
```

```python
width = 0.40

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(np.arange(len(text_x_count)) - width/2, text_x_count.values(), width, label = name_x, color='lightskyblue')
ax.bar(np.arange(len(text_y_count)) + width/2, text_y_count.values(), width, label = name_y)

ax.set_title('Character counts', fontsize=12)

plt.xticks(np.arange(len(text_x_count)), text_x_count.keys(), fontsize=11)
plt.yticks(fontsize=11)
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(np.arange(len(text_x_norm)) - width/2, text_x_norm.values(), width, label = name_x, color='lightskyblue')
ax.bar(np.arange(len(text_y_norm)) + width/2, text_y_norm.values(), width, label = name_y)

ax.set_title('Character probabilities', fontsize=12)

plt.xticks(np.arange(len(text_x_count)), text_x_count.keys(), fontsize=11)
plt.yticks(fontsize=11)

plt.legend()
plt.show()
```
